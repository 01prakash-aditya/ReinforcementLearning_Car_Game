```python
import pandas as pd
import numpy as np 
import tensorflow as tf 
from tensorflow import keras
from keras import layers,activations,optimizers,losses
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from xgboost import XGBClassifier
```

```python
data=pd.read_csv("./time_dilated_train_data.csv",header=None,)
data=data[1:]
data.drop_duplicates(inplace=True)
data.dropna(inplace=True)
for i in range(len(data[7])):
    if(np.array(data[7],dtype="float")[i]==8.0):
        try:
            data.drop(i)
            print("Dropped",i)
        except:
            print("Could not Drop")
data
```

```python
test=data[[7]]
del data[7]
#del data[8]
```

```python

train=data
```

```python
X_train,X_test,y_train,y_test=train_test_split(train,test,test_size=0.2,random_state=112,)
```

```python
X_train=np.asarray(X_train,dtype=float)
X_test=np.asarray(X_test,dtype=float)

y_train=np.asarray(y_train,dtype=float)
y_test=np.asarray(y_test,dtype=float)
```

```python
y_train=np.array([row[0]+1 for row in y_train]).astype("float32")
y_test=np.array([row[0]+1 for row in y_test]).astype("float32")
```

```python
inputshape=X_train[0].shape
outputshape=len(set(y_train))
```

```python
model=keras.Sequential()
model.add(layers.InputLayer(shape=inputshape))
model.add(layers.Dense(128,activation=activations.tanh))
model.add(layers.Dense(256,activation=activations.tanh))
model.add(layers.Dense(outputshape,activation=activations.softmax))

model.compile(loss=losses.sparse_categorical_crossentropy,optimizer=optimizers.Adam(),metrics=["Accuracy"])
model.summary()
```

```python
model.fit(X_train,y_train,epochs=20,batch_size=128)
```

```python
model.evaluate(X_test,y_test)[1]
```

```python
model.save("agentmodel.keras")
```

```python
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
y_pred=xgb.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
